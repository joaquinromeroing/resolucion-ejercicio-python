### Enunciado

En un supermercado de la Ciudad se desea obtener una estadística de las compras realizadas en un período de tiempo en las diferentes sucursales que posee. La información se encuentra almacenada en un archivo COMPRAS cuyo formato de registro es el siguiente (archivo csv):
- SUCURSAL (PRSUC)
- CODIGO DE PRODUCTO (PRCOD)
- FECHA DE COMPRA (PRFEC)
- PROVEEDOR (PRPROV)
- CANTIDAD COMPRADA (PRCANT)
- PRECIO UNITARIO COMPRA (PRPRE)


La información se encuentra ordenada por código de sucursal, producto, fecha de compra y proveedor.
Se debe obtener información sobre:
1. POR PRODUCTO: De cada producto comprado en cada sucursal indicar el total comprado en unidades (TOTUNI) y en pesos (TOTPES)
2. POR SUCURSAL: l informar el total comprado en unidades (TOTSUC) y el producto de mayor (MYPROD, MYIMPOR) y menor compra (MNPRO,MNIMPOR) en pesos.
3. Por otro lado indicar a nivel total: total de sucursales del supermercado (CANSUC) y compra total en pesos de todas las sucursales (TOTALIMP)


Imprimir los resultados obtenidos.

```python
# Ejercicio 3
with open("../data/COMPRAS_supermercado.csv", mode='r') as file:
    next(file)
    line = file.readline()
    data = line.strip().split(",")
    
    prev_suc = data[0]
    cansuc = 1

    totalimp = 0

    while data:

        if prev_suc != data[0]:
            cansuc += 1
        prev_suc = data[0]

        totalimp += int(data[4])*float(data[5])
        
        line = file.readline()
        if not line:
            data = None
        else:
            data = line.strip().split(",")
    
    print(f"Cantida de sucursales: {cansuc}\nImporte total: {round(totalimp,2)}")
```

```python
# Ejercicio 2
with open("../data/COMPRAS_supermercado.csv", mode='r') as file:
    next(file)
    line = file.readline()
    data = line.strip().split(",")

    while data:
        PRSUC = data[0]
        totsuc = 0
        myprod = None
        myimpor = 0
        mnprod = None
        mnimpor = float('inf')

        while data and data[0] == PRSUC:
            PRCOD = data[1]

            totpes = 0

            while data and data[0] == PRSUC and data[1] == PRCOD:
                PRCANT = int(data[4])
                PRPRE  = float(data[5])

                totsuc += PRCANT
                totpes += PRPRE*PRCANT

                line = file.readline()
                if not line:
                    data = None
                else:
                    data = line.strip().split(",")

            if myimpor < totpes:
                myprod = PRCOD
                myimpor = totpes
            
            if mnimpor > totpes:
                mnprod = PRCOD
                mnimpor = totpes
        
        print((f"{PRSUC}\nUnidades Totales: {totsuc}\nMayor importe: {round(myimpor,2)} --> {myprod}\nMenor Importe: {round(mnimpor,2)} --> {mnprod}\n"))
```

```python
# Ejercicio 1
with open("../data/COMPRAS_supermercado.csv", mode='r') as file:
    
    next(file)
    line = file.readline()
    data = line.strip().split(",")
    
    while data:
        PRSUC  = data[0]

        while data and data[0] == PRSUC:
            PRCOD  = data[1]
            
            totuni = 0
            totpes = 0

            while data and data[0] == PRSUC and data[1] == PRCOD:
                PRCANT = int(data[4])
                PRPRE  = float(data[5])

                totuni += PRCANT
                totpes += PRPRE*PRCANT

                line = file.readline()
                if not line:
                    data = None
                else:
                    data = line.strip().split(",")
            
            print(f"{PRSUC}\t{PRCOD}\t{totuni}\t{round(totpes,2)}")
```
