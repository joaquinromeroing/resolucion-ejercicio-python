Enunciado: 

En un supermercado de la Ciudad se desea obtener una estadística de las compras realizadas en
un período de tiempo en las diferentes sucursales que posee. La información se encuentra
almacenada en un archivo COMPRAS cuyo formato de registro es el siguiente (archivo csv):
SUCURSAL (PRSUC)
CODIGO DE PRODUCTO (PRCOD)
FECHA DE COMPRA (PRFEC)
PROVEEDOR (PRPROV)
CANTIDAD COMPRADA (PRCANT)
PRECIO UNITARIO COMPRA (PRPRE)

La información se encuentra ordenada por código de sucursal, producto, fecha de compra y
proveedor.
Se debe obtener información sobre:
a) POR PRODUCTO: De cada producto comprado en cada sucursal indicar el total comprado
en unidades (TOTUNI) y en pesos (TOTPES)
b) POR SUCURSAL: l informar el total comprado en unidades (TOTSUC) y el producto de
mayor (MYPROD, MYIMPOR) y menor compra (MNPRO,MNIMPOR) en pesos.
c) Por otro lado indicar a nivel total: total de sucursales del supermercado (CANSUC) y
compra total en pesos de todas las sucursales (TOTALIMP)

Imprimir los resultados obtenidos.

1-a) POR PRODUCTO: De cada producto comprado en cada sucursal indicar el total comprado
en unidades (TOTUNI) y en pesos (TOTPES)

```python
compras_csv['IMPORTE_TOTAL'] = compras_csv['PRPRE'] * compras_csv['PRCANT']
compras_csv.groupby(['PRSUC','PRCOD']).agg(tot_uni=('PRCANT','sum'), tot_imp=('IMPORTE_TOTAL','sum')).reset_index()
```

```python
i = 0
n = len(compras_csv)
while i < n: 
    Sucursal = compras_csv.loc[i, 'PRSUC']
    Producto = compras_csv.loc[i, 'PRCOD']

    totuni = 0 
    totpeso = 0 
    while i < n and compras_csv.loc[i, 'PRSUC'] == Sucursal and compras_csv.loc[i, 'PRCOD'] == Producto:
        totuni += compras_csv.loc[i, 'PRCANT']
        totpeso += compras_csv.loc[i, 'PRCANT'] * compras_csv.loc[i, 'PRPRE']
        i += 1
    print(f'Sucursal: {Sucursal}, Producto: {Producto}, Total Unidades: {totuni}, Total en pesos: ${totpeso}')

```

b) POR SUCURSAL: l informar el total comprado en unidades (TOTSUC) y el producto de
mayor (MYPROD, MYIMPOR) y menor compra (MNPRO,MNIMPOR) en pesos.

```python
i = 0
n = len(compras_csv)
while i < n:
    Sucursal = compras_csv.loc[i, 'PRSUC']
    totsuc = 0
    MyPROD = 0
    Myimpor = 0
    MnProd = 0
    Mnimpor = 0

    while i < n and compras_csv.loc[i, 'PRSUC'] == Sucursal:
        Producto = compras_csv.loc[i, 'PRCOD']
        totuni = 0
        totimpor = 0

        while i < n and compras_csv.loc[i, 'PRSUC'] == Sucursal and compras_csv.loc[i, 'PRCOD'] == Producto:
            totuni += compras_csv.loc[i, 'PRCANT']
            totimpor += compras_csv.loc[i, 'PRCANT'] * compras_csv.loc[i, 'PRPRE']
            i += 1

        totsuc += totuni

        if totimpor > Myimpor:
            Myimpor = totimpor
            MyPROD = Producto
        elif totimpor < Mnimpor or MnProd == 0:
            Mnimpor = totimpor
            MnProd = Producto

    print(f"Sucursal: {Sucursal}, Total Unidades: {totsuc}")
    print(f"Producto más comprado: {MyPROD} con importe ${Myimpor}")
    print(f"Producto menos comprado: {MnProd} con importe ${Mnimpor}")

```

c) Por otro lado indicar a nivel total: total de sucursales del supermercado (CANSUC) y
compra total en pesos de todas las sucursales (TOTALIMP)

```python
i = 0
n = len(compras_csv)

Cansuc = 0
totalimp = 0
tabla_sucursales = []

while i < n:
    Sucursal = compras_csv.loc[i, 'PRSUC']
    totsucursal = 0
    Cansuc += 1

    while i < n and compras_csv.loc[i, 'PRSUC'] == Sucursal:
        totsucursal += compras_csv.loc[i, 'PRCANT'] * compras_csv.loc[i, 'PRPRE']
        i += 1

    totalimp += totsucursal

    #print(f"Nombre de la sucursal: {Sucursal}, vendió: ${totsucursal}")
    tabla_sucursales.append([Sucursal, totsucursal,totalimp])
df_sucursales = pd.DataFrame(tabla_sucursales, columns=['Sucursal', 'Total en la sucursal', 'Total General'])

df_sucursales['Total en la sucursal'] = df_sucursales['Total en la sucursal'].apply(lambda x: f"${x:,.2f}")
df_sucursales['Total General'] = df_sucursales['Total General'].apply(lambda x: f"${x:,.2f}")
view_sucursales = df_sucursales.head(14)
view_sucursales

#print(f"Total de sucursales: {Cansuc}")
#print(f"Total general: ${totalimp}")
```
